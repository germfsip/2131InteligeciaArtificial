```python
from PIL import Image, ImageChops, ImageEnhance, ImageOps            # funciones para cargar y manipular imágenes
import numpy as np                # funciones numéricas (arrays, matrices, etc.)
import matplotlib.pyplot as plt   # funciones para representación gráfica

I = Image.open("lena.jpg") # Aperturamos la imagen y la guardamos en un objeto I
I.show() # intentamos desplegar la imagen

plt.imshow(np.asarray(I)) #convertimos la imagen en un arreglo de informacion y generamos la gráfica
plt.show()# desplegamos la gráfica
print(I.size, I.mode, I.format)
print(I)
```

```python
# Invertir colores.
new_image = ImageChops.invert(I)
new_image.show()
plt.imshow(np.asarray(new_image))  #Desplegamos imagen
plt.show()
    
# Escala de grises.
new_image = ImageOps.grayscale(I)
plt.imshow(np.asarray(new_image))  #Desplegamos imagen
plt.show()
    
# Resaltar luces.
new_image = ImageEnhance.Brightness(I).enhance(2)
plt.imshow(np.asarray(new_image))  #Desplegamos imagen
plt.show()
    
# Contraste.
new_image = ImageEnhance.Contrast(I).enhance(4)
plt.imshow(np.asarray(new_image))  #Desplegamos imagen
plt.show()
    
 # Espejo.
new_image = ImageOps.mirror(I)
plt.imshow(np.asarray(new_image))  #Desplegamos imagen
plt.show()
    
  # Cambiar tamaño.
new_image = I.resize((320, 240))
plt.imshow(np.asarray(new_image))  #Desplegamos imagen
plt.show()
    
  # Diminuir nitidez.
new_image = ImageEnhance.Sharpness(I).enhance(-4)
plt.imshow(np.asarray(new_image))  #Desplegamos imagen
plt.show()
```

```python
I1 = I.convert('L') # convierte a escala de grises 
plt.imshow(np.asarray(I1))  #convertimos la imagen en un arreglo de informacion y generamos la gráfica
plt.show()   # desplegamos la gráfica
print (I1.size, I1.mode, I1.format)
```
